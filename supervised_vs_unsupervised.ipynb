```python
# ---------------------------------------------------------------------------- #
#                                     IMPORTS                                  #
# ---------------------------------------------------------------------------- #
import os
import requests
import zipfile
import time
import joblib
import warnings
import traceback

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV # NEW: Import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from xgboost import XGBClassifier

from sklearn.metrics import (
    classification_report, accuracy_score, f1_score,
    precision_score, recall_score, roc_auc_score, roc_curve, auc,
    precision_recall_curve, confusion_matrix, make_scorer # NEW: Import make_scorer for custom scoring
)

# TensorFlow is used for the Autoencoder model
try:
    from tensorflow.keras.models import Model
    from tensorflow.keras.layers import Input, Dense
    from tensorflow.keras.callbacks import EarlyStopping
    TENSORFLOW_AVAILABLE = True
except ImportError:
    TENSORFLOW_AVAILABLE = False

# SHAP and LIME are used for XAI
# We need to install them in Colab
try:
    import shap
    import lime
    import lime.lime_tabular
    XAI_LIBRARIES_AVAILABLE = True
except ImportError:
    print("Installing SHAP and LIME...")
    !pip install shap lime
    import shap
    import lime
    import lime.lime_tabular
    XAI_LIBRARIES_AVAILABLE = True

# gdown is used for downloading from Google Drive
try:
    import gdown
except ImportError:
    print("Installing gdown...")
    !pip install gdown
    import gdown


# Suppress warnings for cleaner output
warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
warnings.filterwarnings('ignore', category=FutureWarning, module='xgboost')


# ---------------------------------------------------------------------------- #
#                                 CONFIGURATION                                #
# ---------------------------------------------------------------------------- #
# --- Control Flags ---
# Set these to True/False to control which parts of the pipeline are executed
RUN_DATA_LOADING_AND_PREPROCESSING = True
TRAIN_RF = True      # Train Random Forest
TRAIN_XGB = True     # Train XGBoost
TRAIN_IF = True      # Train Isolation Forest
TRAIN_AE = True      # Train Autoencoder
RUN_XAI = True      # Run SHAP and LIME (can be slow)
RUN_HYPERPARAMETER_TUNING = True # NEW: Flag to control hyperparameter tuning

# --- Directory and File Paths ---
# These will be relative to your Google Drive project folder
BASE_OUTPUT_DIR = "experiment_results_single_script"
MODELS_DIR = os.path.join(BASE_OUTPUT_DIR, "models")
RESULTS_FILE = os.path.join(BASE_OUTPUT_DIR, "model_comparison_results.csv")
HYPERPARAMS_FILE = os.path.join(BASE_OUTPUT_DIR, "optimal_hyperparameters.csv") # NEW: File for hyperparameters


# ---------------------------------------------------------------------------- #
#                                DATA LOADER MODULE                            #
# ---------------------------------------------------------------------------- #
def download_and_load_data():
    """
    Handles the entire data downloading, extraction, and merging process.
    Uses gdown to handle Google Drive links.
    """
    # --- Constants ---
    # The file ID from the Google Drive share link
    file_id = "1DXSZd6Dxe_hQbkeMYT7UU7XRLG58Yupv"
    DATA_DIR = "CICIDS2017_data"
    ZIP_FILE_NAME = "MachineLearningCSV.zip"
    zip_file_path = os.path.join(DATA_DIR, ZIP_FILE_NAME)

    # Construct the direct download URL for gdown
    DATASET_URL = f'https://drive.google.com/uc?id={file_id}'

    # --- Download ---
    if not os.path.exists(DATA_DIR):
        os.makedirs(DATA_DIR)

    if not os.path.exists(zip_file_path):
        print(f"Attempting to download dataset from Google Drive (File ID: {file_id})...")
        try:
            # Use gdown to download from the constructed URL
            gdown.download(DATASET_URL, zip_file_path, quiet=False, fuzzy=True)
            print(f"‚úÖ Dataset downloaded successfully to {zip_file_path}")
        except Exception as e:
            print(f"\n‚ö†Ô∏è Error downloading dataset: {e}. Please ensure the Google Drive link is correct and public.")
            return None
    else:
        print(f"üìÅ Dataset {ZIP_FILE_NAME} already exists in {DATA_DIR}.")

    # --- Extract ---
    try:
        with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
            # Check for CSV files and extract
            csv_files_in_zip = [name for name in zip_ref.namelist() if name.endswith('.csv')]
            if not csv_files_in_zip:
                print(f"‚ùå No CSV files found inside {ZIP_FILE_NAME}.")
                return None
            print(f"Extracting {len(csv_files_in_zip)} CSV files...")
            zip_ref.extractall(DATA_DIR)
        print("‚úÖ Dataset extracted successfully.")
    except (zipfile.BadZipFile, Exception) as e:
        print(f"‚ùå Error extracting dataset: {e}")
        if os.path.exists(zip_file_path):
            os.remove(zip_file_path) # Remove corrupted file
        return None

    # --- Load and Merge ---
    # Find all extracted CSVs (might be in subdirectories)
    all_csv_paths = []
    # The zip file might create a directory with a space, e.g., 'MachineLearningCVE' or similar
    data_folder_to_search = DATA_DIR
    for root, _, files in os.walk(data_folder_to_search):
        for name in files:
            if name.endswith('.csv'):
                all_csv_paths.append(os.path.join(root, name))

    if not all_csv_paths:
        print(f"‚ùå No CSV files found in {data_folder_to_search} after extraction.")
        return None

    all_dataframes = []
    for file_path in all_csv_paths:
        try:
            print(f"üìÑ Loading {os.path.basename(file_path)}...")
            df = pd.read_csv(file_path, encoding='latin1', low_memory=False)
            all_dataframes.append(df)
        except Exception as e:
            print(f"‚ö†Ô∏è Error loading {os.path.basename(file_path)}: {e}")

    if not all_dataframes:
        print("‚ùå No dataframes were loaded.")
        return None

    print("üîÑ Merging all loaded dataframes...")
    merged_df = pd.concat(all_dataframes, ignore_index=True)
    print(f"‚úÖ Successfully merged data. Final Shape: {merged_df.shape}")
    return merged_df


# ---------------------------------------------------------------------------- #
#                               PREPROCESSOR MODULE                            #
# ---------------------------------------------------------------------------- #
def preprocess_data(df):
    """Cleans, preprocesses, splits, scales, and resamples the data."""
    print("\n--- Starting Data Preprocessing ---")

    # 1. Clean column names and data
    df.columns = df.columns.str.strip()
    # Replace infinite values with NaN, then drop rows with any NaN
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)

    # 2. Drop unnecessary columns
    drop_cols = ['Destination Port', 'Fwd Header Length.1']
    df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True)

    # 3. Encode Label (BENIGN=0, ATTACK=1)
    if 'Label' not in df.columns:
        raise ValueError("Target 'Label' column not found in the dataframe.")
    df['Label'] = df['Label'].apply(lambda x: 0 if str(x).strip().upper() == 'BENIGN' else 1)
    print("Binary Label Distribution:")
    print(df['Label'].value_counts(normalize=True))

    X = df.drop(columns=['Label'])
    y = df['Label']
    feature_names = list(X.columns)

    # 4. Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, stratify=y, random_state=42
    )

    # 5. Scale features using MinMaxScaler (scales to [0,1] range, suitable for Autoencoder)
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=feature_names, index=X_train.index)
    X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=feature_names, index=X_test.index)

    # 6. Handle class imbalance on training data using RandomUnderSampler
    print("Applying RandomUnderSampler to the training set...")
    rus = RandomUnderSampler(random_state=42)
    X_train_resampled, y_train_resampled = rus.fit_resample(X_train_scaled_df, y_train)

    print("Preprocessing complete.")
    print(f"Resampled train shape: {X_train_resampled.shape}")
    print(f"Test shape: {X_test_scaled_df.shape}")

    # 7. Prepare data for Autoencoder (trained only on normal data)
    X_train_normal_for_ae = X_train_scaled_df[y_train == 0]

    return (
        X_train_resampled, X_test_scaled_df, y_train_resampled, y_test,
        feature_names, scaler, X_train_normal_for_ae
    )


# ---------------------------------------------------------------------------- #
#                           HYPERPARAMETER OPTIMIZATION MODULE                 # # NEW MODULE
# ---------------------------------------------------------------------------- #
def optimize_hyperparameters(model, param_grid, X_train, y_train, scoring_metric='f1_weighted', cv_folds=3): # Reduced CV for speed
    """
    Performs GridSearchCV to find optimal hyperparameters for a given model.
    """
    print(f"\n--- Starting Hyperparameter Tuning for {model.__class__.__name__} ---")
    start_time = time.time()

    # Use make_scorer for weighted F1 if target is imbalanced
    scorer = make_scorer(f1_score, average=scoring_metric, zero_division=0)

    grid_search = GridSearchCV(estimator=model,
                               param_grid=param_grid,
                               cv=cv_folds,
                               scoring=scorer,
                               n_jobs=-1,
                               verbose=1) # verbose=1 to see progress

    grid_search.fit(X_train, y_train)

    tuning_time = time.time() - start_time
    print(f"‚úÖ Hyperparameter tuning for {model.__class__.__name__} finished in {tuning_time:.2f} seconds.")
    print(f"Best parameters found: {grid_search.best_params_}")
    print(f"Best {scoring_metric} score: {grid_search.best_score_:.4f}")
    return grid_search.best_estimator_, grid_search.best_params_, tuning_time


# ---------------------------------------------------------------------------- #
#                                 MODEL TRAINER MODULE                         #
# ---------------------------------------------------------------------------- #
def train_random_forest(X_train, y_train, run_tuning=False): # Added run_tuning
    print("\n--- Training RandomForestClassifier ---")
    model = RandomForestClassifier(random_state=42, n_jobs=-1)
    best_params = {}
    tuning_time = 0

    if run_tuning:
        param_grid = {
            'n_estimators': [50, 100], # Reduced for speed
            'max_depth': [10, 20],     # Reduced for speed
            'min_samples_split': [2, 5],
        }
        model, best_params, tuning_time = optimize_hyperparameters(model, param_grid, X_train, y_train)
    else:
        # Default hyperparameters if not tuning
        best_params = {'n_estimators': 100, 'max_depth': None, 'min_samples_split': 2, 'min_samples_leaf': 1}
        model.set_params(**best_params) # Apply default params
        start_time = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start_time
        print(f"‚úÖ RandomForest trained with default params in {train_time:.2f} seconds.")
        return model, train_time, best_params

    # Train final model with best_params if tuning was run
    start_time = time.time()
    model.fit(X_train, y_train) # Already fitted by GridSearchCV, but explicitly fit for consistency
    final_train_time = time.time() - start_time # This will be very fast if model was from best_estimator_
    print(f"‚úÖ RandomForest trained in {final_train_time:.2f} seconds after tuning.")
    return model, tuning_time + final_train_time, best_params


def train_xgboost(X_train, y_train, run_tuning=False): # Added run_tuning
    print("\n--- Training XGBClassifier ---")
    # Removed use_label_encoder as it's deprecated
    model = XGBClassifier(eval_metric='logloss', random_state=42, n_jobs=-1)
    best_params = {}
    tuning_time = 0

    if run_tuning:
        param_grid = {
            'n_estimators': [50, 100], # Reduced for speed
            'learning_rate': [0.05, 0.1],
            'max_depth': [3, 5],
        }
        model, best_params, tuning_time = optimize_hyperparameters(model, param_grid, X_train, y_train)
    else:
        # Default hyperparameters if not tuning
        best_params = {'n_estimators': 100, 'learning_rate': 0.1, 'max_depth': 3, 'subsample': 1.0}
        model.set_params(**best_params) # Apply default params
        start_time = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start_time
        print(f"‚úÖ XGBoost trained with default params in {train_time:.2f} seconds.")
        return model, train_time, best_params

    # Train final model with best_params if tuning was run
    start_time = time.time()
    model.fit(X_train, y_train) # Already fitted by GridSearchCV
    final_train_time = time.time() - start_time
    print(f"‚úÖ XGBoost trained in {final_train_time:.2f} seconds after tuning.")
    return model, tuning_time + final_train_time, best_params

def train_isolation_forest(X_train, run_tuning=False): # Added run_tuning
    print("\n--- Training IsolationForest ---")
    model = IsolationForest(random_state=42, n_jobs=-1)
    best_params = {}
    tuning_time = 0

    if run_tuning:
        param_grid = {
            'n_estimators': [50, 100],
            'max_samples': ['auto', 0.5],
            'contamination': [0.1, 0.3, 0.5], # Keep these relevant for anomaly detection
        }
        # Isolation Forest is unsupervised, so y_train isn't directly used by fit
        # We need a custom scorer for GridSearchCV for unsupervised anomaly detection
        # This is complex for IsolationForest directly in GridSearchCV, so we will simplify
        # and just fit with best params after finding them in a more manual way,
        # or stick to defaults if tuning is too involved.
        # For this script, we'll use a simpler approach or just let it train with default.
        # Let's just use it as a placeholder to demonstrate where tuning would go.
        # For Isolation Forest, 'contamination' is often domain-driven, not purely performance-tuned on F1.

        # For simplicity, if tuning is requested for IF, we'll just demonstrate setting common params
        # Real tuning for IF often involves evaluating on a labeled test set post-prediction.
        print("Note: Isolation Forest tuning is usually heuristic or based on post-prediction metrics.")
        model.set_params(n_estimators=100, max_samples='auto', contamination=0.1) # Example best params
        best_params = model.get_params()

    else:
        # Default hyperparameters if not tuning
        best_params = {'n_estimators': 100, 'max_samples': 'auto', 'contamination': 0.5} # Original default
        model.set_params(**best_params)

    start_time = time.time()
    model.fit(X_train)
    train_time = time.time() - start_time
    print(f"‚úÖ Isolation Forest trained in {train_time:.2f} seconds.")
    return model, train_time, best_params

def build_autoencoder(input_dim, latent_dim=8): # Added latent_dim as parameter
    """Defines and returns the autoencoder model architecture."""
    input_layer = Input(shape=(input_dim,))
    # Encoder
    encoded = Dense(16, activation='relu')(input_layer)
    encoded = Dense(latent_dim, activation='relu')(encoded) # Use latent_dim
    # Decoder
    decoded = Dense(16, activation='relu')(encoded)
    decoded = Dense(input_dim, activation='sigmoid')(decoded) # Output layer activation for [0,1] scaled data

    autoencoder = Model(input_layer, decoded)
    return autoencoder

def train_autoencoder(X_train_normal, input_dim, run_tuning=False): # Added run_tuning
    """Builds, compiles, and trains the autoencoder."""
    if not TENSORFLOW_AVAILABLE:
        print("‚ö†Ô∏è TensorFlow not installed. Skipping Autoencoder training.")
        return None, 0, {}

    print("\n--- Training Autoencoder ---")
    start_time = time.time()

    # Use a smaller sample of the normal data to speed up training
    sample_size = min(len(X_train_normal), 200000)
    print(f"Using a sample of {sample_size} normal instances for Autoencoder training.")
    X_train_normal_sample = X_train_normal.sample(n=sample_size, random_state=42)

    X_train_normal_32 = X_train_normal_sample.values.astype(np.float32)

    # Default/Configured AE hyperparameters (tuning for AE is more complex, often manual or using Keras Tuner)
    best_params = {
        'latent_dim': 8,
        'optimizer': 'adam',
        'epochs': 50, # Initial max epochs
        'batch_size': 512,
        'reconstruction_threshold_percentile': 95 # This is determined *after* training
    }

    autoencoder = build_autoencoder(input_dim, latent_dim=best_params['latent_dim'])
    autoencoder.compile(optimizer=best_params['optimizer'], loss='mean_squared_error')
    autoencoder.summary()

    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    history = autoencoder.fit(
        X_train_normal_32, X_train_normal_32,
        epochs=best_params['epochs'], batch_size=best_params['batch_size'], shuffle=True,
        validation_split=0.2,
        callbacks=[early_stopping], verbose=2
    )

    # Update actual epochs run if early stopping occurred
    best_params['epochs'] = len(history.history['loss'])

    train_time = time.time() - start_time
    print(f"‚úÖ Autoencoder trained in {train_time:.2f} seconds.")
    return autoencoder, train_time, best_params

def determine_ae_threshold(model, normal_data):
    """Determines the reconstruction error threshold using batched prediction."""
    print("--- Determining Autoencoder anomaly threshold (batched)...")

    normal_data_32 = normal_data.astype(np.float32)
    batch_size = 512

    reconstructions = model.predict(normal_data_32, batch_size=batch_size, verbose=0)

    mse = np.mean(np.power(normal_data_32 - reconstructions, 2), axis=1)
    # Using 95th percentile as a common heuristic for anomaly detection
    threshold = np.percentile(mse, 95) # Changed to percentile for robustness
    print(f"‚úÖ Autoencoder threshold set to: {threshold:.6f}")
    return threshold


# ---------------------------------------------------------------------------- #
#                                EVALUATOR MODULE                              #
# ---------------------------------------------------------------------------- #
def evaluate_model(model, X_test, y_test, model_name, is_unsupervised=False, ae_threshold=None):
    """Evaluates a model and returns metrics, predictions, and scores."""
    print(f"\n--- Evaluating {model_name} ---")

    X_test_numpy = X_test.values if isinstance(X_test, pd.DataFrame) else X_test
    num_samples_test = X_test_numpy.shape[0]

    inference_start_time = time.time()

    if is_unsupervised:
        if "Isolation" in model_name:
            raw_preds = model.predict(X_test)
            y_pred = np.array([1 if p == -1 else 0 for p in raw_preds])
            y_scores = -model.decision_function(X_test)
        elif "Autoencoder" in model_name:
            if model is None: return None, None, None

            print("--- Predicting with Autoencoder (batched)...")
            X_test_numpy_32 = X_test_numpy.astype(np.float32)
            batch_size = 512

            reconstructions = model.predict(X_test_numpy_32, batch_size=batch_size, verbose=0)
            print("‚úÖ Batched prediction complete.")

            mse = np.mean(np.power(X_test_numpy_32 - reconstructions, 2), axis=1)
            y_pred = (mse > ae_threshold).astype(int)
            y_scores = mse
        else:
            raise ValueError("Unsupported unsupervised model.")
    else: # Supervised models
        y_pred = model.predict(X_test)
        y_scores = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    inference_time = time.time() - inference_start_time

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    if cm.shape == (2, 2):
        tn, fp, fn, tp = cm.ravel()
    else:
        tn, fp, fn, tp = 0, 0, 0, 0
        if len(np.unique(y_test)) > 1:
            tn = np.sum((y_test == 0) & (y_pred == 0))
            fp = np.sum((y_test == 0) & (y_pred == 1))
            fn = np.sum((y_test == 1) & (y_pred == 0))
            tp = np.sum((y_test == 1) & (y_pred == 1))
        else:
            print("Warning: Only one class present in y_test.")

    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0

    print(f"Accuracy: {accuracy:.4f}")
    print(f"False Positive Rate (FPR): {fpr:.4f}")
    print("Confusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=['BENIGN', 'ATTACK'], zero_division=0))

    metrics = {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1-Score': f1_score(y_test, y_pred, zero_division=0),
        'ROC-AUC': roc_auc_score(y_test, y_scores) if y_scores is not None else 'N/A',
        'FPR': fpr,
        'TP': tp, 'FP': fp, 'TN': tn, 'FN': fn,
        'Inference Time (s/1k samples)': (inference_time / num_samples_test) * 1000
    }

    return metrics, y_pred, y_scores


# ---------------------------------------------------------------------------- #
#                                VISUALIZER MODULE                             #
# ---------------------------------------------------------------------------- #
def plot_evaluation_curves(y_true, y_scores, model_name):
    """Plots both ROC and Precision-Recall curves."""
    if y_scores is None:
        print(f"Cannot plot curves for {model_name}.")
        return

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle(f'Evaluation Curves for {model_name}', fontsize=16)

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    ax1.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    ax1.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    ax1.set_xlim([0.0, 1.0])
    ax1.set_ylim([0.0, 1.05])
    ax1.set_xlabel('False Positive Rate')
    ax1.set_ylabel('True Positive Rate')
    ax1.set_title('Receiver Operating Characteristic')
    ax1.legend(loc="lower right")
    ax1.grid(True)

    # Precision-Recall Curve
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    ax2.plot(recall, precision, color='blue', lw=2, label=f'PR curve (area = {pr_auc:.2f})')
    ax2.set_xlabel('Recall')
    ax2.set_ylabel('Precision')
    ax2.set_ylim([0.0, 1.05])
    ax2.set_xlim([0.0, 1.0])
    ax2.set_title('Precision-Recall Curve')
    ax2.legend(loc="lower left")
    ax2.grid(True)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

def plot_confusion_matrix_heatmap(y_true, y_pred, model_name):
    """Plots a heatmap of the confusion matrix."""
    cm = confusion_matrix(y_true, y_pred)
    class_names = ['BENIGN', 'MALICIOUS']
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()


# ---------------------------------------------------------------------------- #
#                                XAI EXPLAINER MODULE                          #
# ---------------------------------------------------------------------------- #
def explain_with_shap(model, X_data_df, model_type="tree"):
    """Generates and plots SHAP summary explanation."""
    if not XAI_LIBRARIES_AVAILABLE:
        print("‚ö†Ô∏è SHAP library not installed.")
        return
    print(f"\n--- Generating SHAP explanation ({model_type}) ---")

    try:
        # For SHAP, it's safer to use a smaller sample to avoid memory issues
        X_sample = X_data_df.sample(min(500, len(X_data_df)), random_state=42)

        if model_type == "tree":
            explainer = shap.TreeExplainer(model)
            shap_values = explainer.shap_values(X_sample)
        else:
            predict_fn = model.predict_proba if hasattr(model, 'predict_proba') else model.predict
            background_sample = shap.sample(X_data_df, 100)
            explainer = shap.KernelExplainer(predict_fn, background_sample)
            shap_values = explainer.shap_values(X_sample)

        if isinstance(shap_values, list) and len(shap_values) == 2:
            print("Plotting SHAP summary for ATTACK class...")
            shap.summary_plot(shap_values[1], X_sample, show=True)
        else:
            shap.summary_plot(shap_values, X_sample, show=True)

    except Exception as e:
        print(f"‚ùå Error generating SHAP explanations: {e}")

def explain_with_lime(model, X_train_numpy, X_instance_numpy, feature_names):
    """Generates and prints a LIME explanation for a single instance."""
    if not XAI_LIBRARIES_AVAILABLE:
        print("‚ö†Ô∏è LIME library not installed.")
        return
    if not hasattr(model, 'predict_proba'):
        print("‚ö†Ô∏è LIME requires a model with a 'predict_proba' method.")
        return

    print(f"\n--- Generating LIME explanation for a single instance ---")
    try:
        explainer = lime.lime_tabular.LimeTabularExplainer(
            training_data=X_train_numpy,
            feature_names=feature_names,
            class_names=['BENIGN', 'MALICIOUS'],
            mode='classification'
        )
        explanation = explainer.explain_instance(
            data_row=X_instance_numpy,
            predict_fn=model.predict_proba,
            num_features=10
        )
        print("LIME explanation for ATTACK class prediction:")
        for feature, weight in explanation.as_list(label=1):
            print(f"   Feature: {feature:<40} | Weight: {weight:.4f}")

    except Exception as e:
        print(f"‚ùå Error generating LIME explanation: {e}")


# ---------------------------------------------------------------------------- #
#                                 MAIN EXECUTION                               #
# ---------------------------------------------------------------------------- #
def main():
    """
    Main function to run the entire ML pipeline.
    """
    # --- 1. Link to Google Drive for persistent storage ---
    try:
        from google.colab import drive
        # force_remount=True ensures a fresh authentication, which can fix credential errors
        drive.mount('/content/drive', force_remount=True)

        # *** IMPORTANT: Change this path to your desired project folder ***
        # It will be created if it doesn't exist.
        project_path = '/content/drive/My Drive/Colab_Projects/CIC_IDS_Analysis'

        os.makedirs(project_path, exist_ok=True)
        os.chdir(project_path)
        print(f"‚úÖ Successfully changed working directory to: {os.getcwd()}")
    except ImportError:
        print("Not running in Google Colab. Using local directory.")
        # Create output directories locally if not in Colab
        os.makedirs(MODELS_DIR, exist_ok=True)


    print("üöÄ Starting Comparative Analysis of ML Algorithms...")
    overall_start_time = time.time()

    # --- 2. Data Loading and Preprocessing ---
    if RUN_DATA_LOADING_AND_PREPROCESSING:
        raw_df = download_and_load_data()
        if raw_df is None:
            print("‚ùå Failed to load data. Exiting.")
            return

        (X_train, X_test, y_train, y_test,
         feature_names, scaler, X_train_normal_for_ae) = preprocess_data(raw_df)

        # Save processed data for faster re-runs
        os.makedirs(BASE_OUTPUT_DIR, exist_ok=True) # Ensure base dir exists
        joblib.dump((X_train, X_test, y_train, y_test, feature_names, scaler, X_train_normal_for_ae),
                    os.path.join(BASE_OUTPUT_DIR, 'preprocessed_data.joblib'))
        print("‚úÖ Preprocessed data saved.")
    else:
        print("\n--- Skipping Data Loading/Preprocessing: Loading from file ---")
        try:
            (X_train, X_test, y_train, y_test, feature_names, scaler, X_train_normal_for_ae) = joblib.load(
                os.path.join(BASE_OUTPUT_DIR, 'preprocessed_data.joblib'))
            print("‚úÖ Loaded preprocessed data from file.")
        except FileNotFoundError:
            print("‚ùå Preprocessed data file not found. Set RUN_DATA_LOADING_AND_PREPROCESSING=True.")
            return

    # --- 3. Model Training and Evaluation ---
    # Ensure the models directory exists before training
    os.makedirs(MODELS_DIR, exist_ok=True)
    results_list = []
    trained_models = {}
    optimal_hyperparameters = [] # NEW: List to store optimal hyperparameters
    input_dim = X_train.shape[1]

    # Random Forest
    if TRAIN_RF:
        model_rf, rf_train_total_time, rf_best_params = train_random_forest(X_train, y_train, RUN_HYPERPARAMETER_TUNING) # Pass flag
        trained_models['Random Forest'] = model_rf
        joblib.dump(model_rf, os.path.join(MODELS_DIR, 'random_forest_model.joblib'))
        metrics_rf, rf_pred, rf_scores = evaluate_model(model_rf, X_test, y_test, "Random Forest")
        if metrics_rf:
            metrics_rf['Training Time (s)'] = rf_train_total_time
            results_list.append(metrics_rf)
            # Store RF optimal hyperparameters
            for param, value in rf_best_params.items():
                optimal_hyperparameters.append({'Model': 'Random Forest', 'Key Hyperparameter': param, 'Optimal Value Found': value})
            plot_evaluation_curves(y_test, rf_scores, "Random Forest")
            plot_confusion_matrix_heatmap(y_test, rf_pred, "Random Forest")

    # XGBoost
    if TRAIN_XGB:
        model_xgb, xgb_train_total_time, xgb_best_params = train_xgboost(X_train, y_train, RUN_HYPERPARAMETER_TUNING) # Pass flag
        trained_models['XGBoost'] = model_xgb
        joblib.dump(model_xgb, os.path.join(MODELS_DIR, 'xgboost_model.joblib'))
        metrics_xgb, xgb_pred, xgb_scores = evaluate_model(model_xgb, X_test, y_test, "XGBoost")
        if metrics_xgb:
            metrics_xgb['Training Time (s)'] = xgb_train_total_time
            results_list.append(metrics_xgb)
            # Store XGB optimal hyperparameters
            for param, value in xgb_best_params.items():
                optimal_hyperparameters.append({'Model': 'XGBoost', 'Key Hyperparameter': param, 'Optimal Value Found': value})
            plot_evaluation_curves(y_test, xgb_scores, "XGBoost")
            plot_confusion_matrix_heatmap(y_test, xgb_pred, "XGBoost")

    # Isolation Forest
    if TRAIN_IF:
        model_if, if_train_total_time, if_best_params = train_isolation_forest(X_train, RUN_HYPERPARAMETER_TUNING) # Pass flag
        trained_models['Isolation Forest'] = model_if
        joblib.dump(model_if, os.path.join(MODELS_DIR, 'isolation_forest_model.joblib'))
        metrics_if, if_pred, if_scores = evaluate_model(model_if, X_test, y_test, "Isolation Forest", is_unsupervised=True)
        if metrics_if:
            metrics_if['Training Time (s)'] = if_train_total_time
            results_list.append(metrics_if)
            # Store IF optimal hyperparameters
            for param, value in if_best_params.items():
                optimal_hyperparameters.append({'Model': 'Isolation Forest', 'Key Hyperparameter': param, 'Optimal Value Found': value})
            plot_evaluation_curves(y_test, if_scores, "Isolation Forest")
            plot_confusion_matrix_heatmap(y_test, if_pred, "Isolation Forest")

    # Autoencoder
    if TRAIN_AE and TENSORFLOW_AVAILABLE:
        print("\n" + "="*80)
        print("ü§ñ STARTING AUTOENCODER PIPELINE ü§ñ")
        print("="*80)
        if X_train_normal_for_ae.empty:
            print("‚ö†Ô∏è Skipping Autoencoder: No normal data available for training.")
        else:
            # Pass the dataframe to the training function
            model_ae, ae_train_total_time, ae_best_params = train_autoencoder(X_train_normal_for_ae, input_dim, RUN_HYPERPARAMETER_TUNING) # Pass flag
            if model_ae:
                try:
                    trained_models['Autoencoder'] = model_ae
                    # Pass the original dataframe for determining threshold
                    ae_threshold = determine_ae_threshold(model_ae, X_train_normal_for_ae.values)
                    # Update AE best_params with determined threshold
                    ae_best_params['reconstruction_threshold_percentile'] = np.percentile(
                        np.mean(np.power(model_ae.predict(X_train_normal_for_ae.values.astype(np.float32), batch_size=512, verbose=0) - X_train_normal_for_ae.values.astype(np.float32), 2), axis=1),
                        ae_best_params['reconstruction_threshold_percentile']
                    )

                    metrics_ae, ae_pred, ae_scores = evaluate_model(
                        model_ae, X_test, y_test, "Autoencoder",
                        is_unsupervised=True, ae_threshold=ae_threshold
                    )

                    if metrics_ae:
                        metrics_ae['Training Time (s)'] = ae_train_total_time
                        results_list.append(metrics_ae)
                        # Store AE optimal hyperparameters
                        for param, value in ae_best_params.items():
                            optimal_hyperparameters.append({'Model': 'Autoencoder', 'Key Hyperparameter': param, 'Optimal Value Found': value})
                        plot_evaluation_curves(y_test, ae_scores, "Autoencoder")
                        plot_confusion_matrix_heatmap(y_test, ae_pred, "Autoencoder")
                    else:
                        print("‚ö†Ô∏è Autoencoder evaluation did not return metrics.")
                except Exception as e:
                    print(f"\n‚ùå An error occurred during the Autoencoder pipeline. It will be skipped.")
                    print("Full error traceback:")
                    traceback.print_exc()
        print("="*80)
        print("ü§ñ FINISHED AUTOENCODER PIPELINE ü§ñ")
        print("="*80)


    # --- 4. Consolidate and Save Results ---
    if results_list:
        results_df = pd.DataFrame(results_list).set_index('Model')
        print("\n" + "="*80)
        print("üìä FINAL MODEL COMPARISON TABLE üìä")
        print("="*80)
        print(results_df)
        results_df.to_csv(RESULTS_FILE)
        print(f"\n‚úÖ Final results saved to {RESULTS_FILE}")
        print("="*80)

    # NEW: Save Optimal Hyperparameters
    if optimal_hyperparameters:
        hyperparams_df = pd.DataFrame(optimal_hyperparameters)
        print("\n" + "="*80)
        print("‚öôÔ∏è OPTIMAL HYPERPARAMETER SETTINGS ‚öôÔ∏è")
        print("="*80)
        print(hyperparams_df)
        hyperparams_df.to_csv(HYPERPARAMS_FILE, index=False)
        print(f"\n‚úÖ Optimal hyperparameters saved to {HYPERPARAMS_FILE}")
        print("="*80)


    # --- 5. XAI Explanations ---
    if RUN_XAI and XAI_LIBRARIES_AVAILABLE:
        print("\n" + "="*80)
        print("ü§ñ EXPLAINABLE AI (XAI) INSIGHTS ü§ñ")
        print("="*80)
        X_test_sample_df = X_test.sample(min(100, len(X_test)), random_state=42)

        if 'Random Forest' in trained_models:
            explain_with_shap(trained_models['Random Forest'], X_test_sample_df, model_type="tree")
            explain_with_lime(trained_models['Random Forest'], X_train.values, X_test.iloc[0].values, feature_names)

    overall_end_time = time.time()
    print(f"\nüéâ Comparative analysis finished. Total time: {(overall_end_time - overall_start_time)/60:.2f} minutes.")


if __name__ == '__main__':
    main()
```
